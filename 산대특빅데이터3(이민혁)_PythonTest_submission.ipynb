#### Question3
아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오. 

```python
list1 = [1, 50, 27, 19, 37, 33, 5, 3, 48]

even, odd = 0, 0

for i in list1 :
    if i % 2 == 0 :
        even += 1
    else :
        odd += 1
print(f'짝수 {even}개, 홀수 {odd}개')
```

#### Question4
다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 프로그램을 작성하시오. (단, 학생의 숫자가 늘어나도 동작할 수 있도록 고려하시오.)

```python
score_list = [['김미희',89,78,99],['서대희', 67,88,77],['강예진',100,55,96],['차연석',75,100,60]]

grade = ''
p,j,d = 0,0,0

print('==============================================')
print('이름    Python   Java     DB    평균    학점')
print('==============================================')

for i in range(0,len(score_list)) :
    python, java, db = score_list[i][1:]
    total = python + java + db
    avg = int(total / 3)
    p += python
    j += java
    d += db
    if avg > 90 :
        grade = 'A'
    elif avg > 80 :
        grade = 'B'
    elif avg > 70 :
        grade = 'C'
    elif avg > 60 :
        grade = 'D'
    else :
        grade = 'F'
    print(score_list[i][0],'\t',python,'\t',java,'\t',db,'\t',avg,'\t',grade)
print('==============================================')
print(f'평균     {int(p/len(score_list))}      {int(j/len(score_list))}      {int(d/len(score_list))}')
```

※ 조건 내용 반드시 확인할 것  
 조건 1. 사고유형_대분류가 ‘차대사람’ 이면서 발생시간대가 ‘야간’인 경우의 사고 건수 카운트  
 조건 2. y축의 범위는 0~200으로 설정할 것  
 조건 3. 차트 내부 글씨는 한글로 표시할 것  

```python
import pandas as pd

data = pd.read_csv('./data/Traffic_Accident_2017.csv', encoding='euc-kr')
```

```python
data.tail(10)
```

```python
mon = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','월','보행자']].value_counts()
tue = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','화','보행자']].value_counts()
wen = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','수','보행자']].value_counts()
thr = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','목','보행자']].value_counts()
fri = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','금','보행자']].value_counts()
sat = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','토','보행자']].value_counts()
sun = data.loc[:,['주야','요일','당사자종별_2당_대분류']][data.loc[:,['주야','요일','당사자종별_2당_대분류']] == ['야간','일','보행자']].value_counts()
```

```python
data['요일']
```

```python
y = data['요일'].value_counts()
```

```python
y = y[['월','화','수','목','금','토','일']]
```

```python
from matplotlib import rc

rc('font', family = 'Malgun Gothic')
```

```python
x = y.index
acc = [mon[0],tue[0],wen[0],thr[0],fri[0],sat[0],sun[0]]
plt.bar(x,acc)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.ylim(0,200)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.show()
```
